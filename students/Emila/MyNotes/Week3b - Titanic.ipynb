```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot
import sklearn.linear_model
```

```python
titanic = pd.read_csv('../../../data/titanic.csv', index_col='PassengerId')
titanic
```

```python
#get rid of all rows where passenger has no age
p_with_age = titanic[titanic.Age.notnull()]
```

```python
age = p_with_age[['Age']]
```

```python
regressor = sklearn.linear_model.LogisticRegression()
```

```python
regressor.fit(age, p_with_age.Survived)
```

```python
matplotlib.pyplot.scatter(p_with_age.Age, p_with_age.Survived)
matplotlib.pyplot.scatter(p_with_age.Age, regressor.predict(age), color='red')
```

```python
regressor.coef_
#based on your age, there's a 1% relationship, the older you got the more likely you'll die
#1 yr old had 10% more change of survival than 100 year old
```

```python
#regress using class & gender
#first turn gender into a number
import numpy
```

```python
#numpy.where(titanic.Sex == 'male', 0, 1)
#can also do this which returns a pandas series
titanic['gender'] = titanic.Sex.map({'male':0, 'female':1})
```

```python
X = titanic[['gender', 'Pclass']]
y = titanic.Survived
```

```python
r2 = sklearn.linear_model.LogisticRegression()
r2.fit(X, y)
```

```python
#predict probability of female in 1st class
r2.predict_proba([[1,1]])
```

```python
#predict probability of male in 1st class
r2.predict_proba([[0,1]])
```

```python
#predict probability of male in 3rd class
r2.predict_proba([[0,3]])
```

```python
#predict probability of female in 3rd class
r2.predict_proba([[1,3]])
```

```python
predicted_survival = r2.predict(X)
```
