```python
#auto-correlation
import statsmodels.api
```

```python
#lags = 100 limit to 2 years
statsmodels.api.graphics.tsa.plot_acf(syd_melb.Passenger_Trips, lags=24)
```

```python
#given that we already know it's always correlated to the previous month... let's get the correlation again:
statsmodels.api.graphics.tsa.plot_pacf(syd_melb.Passenger_Trips, lags=24)
#this shows... normally things are quite correlated
#12th month is more corrleated, but 13 month is less so??
#if you took a domestic flight 13 month ago you're much less likely to take one now.
```

```python
#what do these graphs mean?
statsmodels.api.graphics.tsa.month_plot(syd_melb.Passenger_Trips)
```

```python
#freq=12 means I have 1 datapoint per month
dcr = statsmodels.api.tsa.seasonal_decompose(syd_melb.Passenger_Trips, freq = 12, model='multiplicative')
```

```python
#manually detrending
detrend = statsmodels.api.tsa.detrend(syd_melb.Passenger_Trips)
```

```python
import matplotlib
#what does the detrended graph look like? Ok we'll still need to take out the seasonal stuff..
matplotlib.pyplot.plot(detrend)
```

```python
trend = syd_melb.Passenger_Trips - detrend
```

```python
#can linearly regress this!
trend.plot()
```

```python
#get me the mean of all the january data
detrend[0::12].mean()
```

```python
detrend[9::12].mean()
```

```python
monthly_trends = []
for i in range(12):
    monthly_trends.append(detrend[i::12].mean())
    
#also can use: monthly_trends = [detrend[i::12] for i in range(12)]
```

```python
matplotlib.pyplot.plot(monthly_trends)
```

```python
#need to pull seasonsal trend out of my data now:
#extend & replicate monthly_trends to the length of detrend graph
deseasoned = detrend - (monthly_trends * 50)[:len(detrend)]
```

```python
matplotlib.pyplot.plot(deseasoned)
#made it slightly narrower.. now I can do my ARMA model
```

```python
#even having extracted all this stuff out it still makes sense to go 12 back?
statsmodels.api.graphics.tsa.plot_acf(deseasoned, lags = 40)
```

```python
statsmodels.api.graphics.tsa.plot_pacf(deseasoned, lags = 40)
#ok that doesn't look as strong
```

```python
#now use ARIMA on deseasoned
a =statsmodels.api.tsa.ARIMA(deseasoned, (12, 0, 3))
```
