```python
#this looks like it will be multiplicative
holtwinters.multiplicative(
    list(syd_melb.Aircraft_Trips),
    #seasonal should be 12
    12,
    #project forwards 2 years
    24,
)
#gives you projection for the next 2 years:
```

```python
#put it in a variable
prediction = holtwinters.multiplicative(
    #chop of 2 years of data
    list(syd_melb.Aircraft_Trips)[:-24],
    #seasonal should be 12
    12,
    #add few extra years
    48,
)
prediction
```

```python
#assign the variables data, level_decay, trend_decay, seasonal_decay, root_mse to the 4 outputs of the prediction
#root_mse is the error
(data, level_decay, trend_decay, seasonal_decay, root_mse) = prediction
len(data)
```

```python
pd.date_range('2014-05-01', periods=48, freq='M')
```

```python
#make a series out of the new predicted data
predicted_aircraft_trips = pd.Series(data=data, 
                                     index=pd.date_range('2014-05-01', periods=48, freq='M'))
```

```python
#plot the original graph and the predicted data:
syd_melb.Aircraft_Trips.plot()
predicted_aircraft_trips.plot(c='red')
```
