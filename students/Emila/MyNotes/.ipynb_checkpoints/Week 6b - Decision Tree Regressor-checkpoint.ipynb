## Decision Tree Regressor
Goal: Predict income from phone case colour and age

```python
import pandas as pd
import numpy
```

```python
phonecases = pd.read_csv('../../../data/phonecase.csv')
```

```python
set(phonecases.Colour)
```

```python
for colour in set(phonecases.Colour):
    phonecases[colour] = numpy.where(phonecases.Colour == colour,1,0)
```

```python
phonecases
```

```python
X = phonecases[['Red', 'Pink', 'Yellow', 'Age']]
```

```python
y = phonecases.Income
```

```python
import sklearn.tree
```

```python
regressor = sklearn.tree.DecisionTreeRegressor(max_depth=3)
```

```python
regressor.fit(X,y)
```

```python
#funky python function zip puts two arrays together
zip(regressor.predict(X), y)
```

```python
import StringIO
s = StringIO.StringIO()
sklearn.tree.export_graphviz(regressor, out_file = s, feature_names=['Red', 'Pink', 'Yellow', 'Age'])
```

```python
import pydotplus
graph = pydotplus.graph_from_dot_data(s.getvalue())
```

```python
import IPython.display
IPython.display.Image(graph.create_png())
```

```python
import sklearn.grid_search
```

```python
dtr = sklearn.tree.DecisionTreeRegressor()
brute_force = sklearn.grid_search.GridSearchCV(dtr, {'max_depth': range(1,10)}, cv=6, scoring='mean_squared_error')
```

```python
brute_force.fit(X, y)
```

```python
brute_force.best_params_
```

```python
s = StringIO.StringIO()
sklearn.tree.export_graphviz(brute_force.best_estimator_, out_file = s, feature_names=['Red', 'Pink', 'Yellow', 'Age'])
graph = pydotplus.graph_from_dot_data(s.getvalue())
IPython.display.Image(graph.create_png())
```

## Bagging
Using the housing data to see the effects of ensemble/bagging

```python
import sklearn.datasets
```

```python
# this is a dictionary
b = sklearn.datasets.load_boston()
```

```python
b.keys()
```

```python
b['feature_names']
```

```python
#describe the dataset
print b['DESCR']
```

```python
Boston = pd.DataFrame(data=b['data'], columns=b['feature_names'])
```

```python
Boston['price'] = b['target']
Boston.head()
```

```python
import sklearn.ensemble
```

```python
#Use regressor because we're trying to predict a number
#Make 25 decision trees on 100 rows, using only 3 columns each. May have overlapping features
#Want to have a reasonable amount of overlap but not too much.
bag = sklearn.ensemble.BaggingRegressor(max_features=3, n_estimators=25, max_samples=100, n_jobs=-1)
```

```python
#Create a dataframe
boston_predictors = Boston.drop('price', axis=1)
#also can do: 
#Boston[Boston.columns[:-1]]
bag.fit(boston_predictors, Boston.price)
```

```python
#Here are all the 25 decision trees that were used
bag.estimators_
```

```python
s = StringIO.StringIO()
sklearn.tree.export_graphviz(bag.estimators_[24], out_file = s, feature_names=Boston.columns)
graph = pydotplus.graph_from_dot_data(s.getvalue())
IPython.display.Image(graph.create_png())
```

```python
zip(bag.predict(boston_predictors), Boston.price)
```

```python
%matplotlib inline
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(Boston.price, bag.predict(boston_predictors))
#Underestimates quite consistantly, doing quite well until 50k?
#We can regress this line and see how much it can explain TheilSen or Ransac
#can do a grid search on this to try some different parameters
```

## Random Forest

```python
# need a lot more than 25, because each tree is so bad!
rdf = sklearn.ensemble.RandomForestRegressor(n_estimators=100)
```

```python
rdf.fit(boston_predictors, Boston.price)
```

```python
plt.scatter(Boston.price, rdf.predict(boston_predictors))
```

```python
zip(rdf.feature_importances_, Boston.columns)
#LSTAT(how many people with low income) and RM(average rooms) are particularly strong,
#represents the strong line in the graph when estimator is 1 (it should normally be super scattered)
```
